```python
#importando pandas
import pandas as pd

#asignamos lectura de archivo csv a variable
inmuebles = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Inmersion de Datos con Python -Alura Latam/inmuebles_bogota.csv")
```

```python
#Quitando la tilde o acento del Área para mayor facilidad
columnas = { "Área": "Area"}
inmuebles = inmuebles.rename(columns = columnas)
#Probando 
inmuebles.sample()
```

1. Promedio de área de inmuebles




```python
#promedio de área de los inmuebles en dataset general
promedio= inmuebles.Area.mean()
print(f"El promedio general de área de los inmuebles en venta en Bogota es:{promedio.round(2)}")
```

1. Promedio de área de todos los inmuebles en los barrios en el dataset. El top 10 del promedio del dataset.

```python
#Promedio de area de inmuebles por barrio
promedio_area_por_barrio = inmuebles.groupby(["Barrio"]).Area.mean()
#Top 10 del Promedio de area de inmuebles por barrio
promedio_area_por_barrio_descendente = promedio_area_por_barrio.sort_values(ascending=False).head(10)
#Imprimiendo 
promedio_area_por_barrio_descendente
```

```python
#Otra forma resumida del punto anterior
inmuebles.groupby(["Barrio"]).Area.mean().sort_values(ascending=False).head(10)
```

```python
#graficando
promedio_area_por_barrio_descendente.plot.bar()
```

> 2.1. Datos estadisticos basicos con el metodo describe() en el dataframe general y por barrios




```python
#Datos estadistios basicos: Conteo, promedio, mediana o segundo uarti, valor minimo y valor maximo del área de todo el dataset
inmuebles.Area.describe()
```

```python
#Datos estadistios basicos: Conteo, promedio, mediana o segundo uarti, valor minimo y valor maximo del área por barrio
estadisticos_area_barrio= inmuebles.groupby(["Barrio"]).Area.describe()
estadisticos_area_barrio
```

```python
#Datos estadistios basicos: Conteo, promedio, mediana o segundo quartil, valor minimo y valor maximo del área en top 10 por barrio
estadisticos_area_barrio= inmuebles.groupby(["Barrio"]).Area.describe()
top_area_estadisticos= estadisticos_area_barrio.sort_values("mean", ascending=False).head(10)
top_area_estadisticos
```

```python
#Verificando 
#promedio de área de los inmuebles en dataset general
promedio= inmuebles.Area.mean()
print(f"El promedio de área de todos los inmuebles en venta en Bogota es:{promedio.round(2)}")
#mediana
mediana= inmuebles.Area.median()
print(f"La Mediana de área de todos los inmuebles en venta en Bogota es:{mediana.round(2)}")
#conteo
conteo= inmuebles.Area.count()
print(f"La cantidad de de inmuebles en venta en Bogota es:{conteo.round(2)}")
#valor minimo
vmin = inmuebles.Area.min()
print(f"El áres minima de los inmuebles en venta en Bogota es: {vmin.round()} metros cuadrados ")
#valor maximo
vmax = inmuebles.Area.max()
print(f"El áres maxima de los inmuebles en venta en Bogota es: {vmax.round()} metros cuadrados")
```

```python
#Encontrando valor y Barrio de inmueble en venta más pequeño de Bogota
index_vmin= inmuebles.Area.idxmin()
print(index_vmin)
valor_inmueble_mas_pqno = inmuebles["Valor"][index_vmin]
barrio_inmueble_mas_pqno=  inmuebles["Barrio"][index_vmin]
print(f"El valor del inmueble más pequeño de Bogota es de {valor_inmueble_mas_pqno} pesos colombianos y se encuentra ubicado en el barrio {barrio_inmueble_mas_pqno}, sin embargo, es posible que el área de este inmueble presente un error en la digitación.")
```

```python
#Encontrando valor y barrio de inmueble  en venta más grande de Bogota
index_vmax= inmuebles.Area.idxmax()
print(index_vmax)
valor_inmueble_mas_grande= inmuebles["Valor"][index_vmax]
barrio_inmueble_mas_grande= inmuebles["Barrio"][index_vmax]
print(f"El valor del inmueble más grande de Bogota es de {valor_inmueble_mas_grande} pesos colombianos y se encuentra ubicado en el barrio {barrio_inmueble_mas_grande}")
```
